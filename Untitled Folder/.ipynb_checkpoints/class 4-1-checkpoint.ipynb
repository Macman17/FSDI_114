```python
class DNode:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None
        
class DoublyLinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, value):
        if not self.head:
            self.head = DNode(value)
        else:
            current = self.head
            while current.next:
                current = current.next
            new_node = DNode(value)
            current.next = new_node
            new_node.prev = current
    
    def __str__(self):
        out = "[%s" % self.head.data
        current = self.head.next
        while current:
            out += ", %s" % current.data
            current = current.next
        out += "]"
        return out
    #method that all us to allow us to print our list backwards
    def get_tail(self):
        current = self.head
        while current.next:
            current = current.next
        return current  
    
    def __len__(self):
        counter = 0 
        current = self.head
        while current:
            current = current.next
            counter += 1
        return counter    
    
    def print_inverted(self):
            current = self.get_tail()
            out = "[%s" % current.data
            current = current.prev
            while current:
                out += ", %s" % current.data
                current = current.prev
            out += "]"
            return out
            
    #method that takes parameters: index and value
    def insert(self, index, value):
        if index < 0:
            raise IndexError(" Negative index not supported.") 
        new_node = DNode(value)
        if index == 0:
            temp = self.head
            self.head = new_node
            new_node.next = temp
        
        else:
            current = self.head
            counter = 0
            while current.next and counter != index: #this handle the anywhere in the middle

                current = current.next
                counter += 1                        
            if index > counter:
                current.next = new_node
                new_node.prev = current
            elif index == counter:
                prev_node = current.prev
                prev_node.next = new_node
                new_node.prev = prev_node
                new_node.next = current
                current.prev = new_node
```

```python
dll = DoublyLinkedList()

for i  in range(5):
    dll.append(i)
    
print(dll)

print("Our list after insert the head:")
dll.insert(2,22)
print(dll)
    
print(dll.print_inverted())
```

```python
from functools import lru_cache 

@lru_cache
def fib(n):
    if n < 2:
        return n 
    return fib(n-1) + fib(n-2)


def fib_n(n):
    if n < 2:
        return n 
    return fib_n(n-1) + fib_n(n-2)
```

```python
my_cache = {}

def fib_d(n):
    val = my_cache.get(n)
    if val:
        return val
    if n < 2:
        return n 
    my_cache[n-1] = fib_d(n-1)
    my_cache[n-2] = fib_d(n-2)
    return my_cache[n-1] + my_cache[n-2]
    
```
